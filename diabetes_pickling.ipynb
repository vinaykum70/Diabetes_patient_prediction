## Data Preprocessing

### Outliers Treatment

```python
from feature_engine.outliers import Winsorizer

win = Winsorizer(capping_method='iqr', tail='both', fold=1.5, variables=['BloodPressure','Insulin','BMI','DiabetesPedigreeFunction','Age'])

df_win = win.fit_transform(df[['BloodPressure','Insulin','BMI','DiabetesPedigreeFunction','Age']])

df_win
```

```python
df['BloodPressure'] = df_win['BloodPressure']

df['Insulin'] = df_win['Insulin']

df['BMI'] = df_win['BMI']

df['DiabetesPedigreeFunction'] = df_win['DiabetesPedigreeFunction']

df['Age'] = df_win['Age']
```

```python
df.head()
```

### X & y

```python
X = df.drop('Diabetes', axis=1)

y = df['Diabetes']
```

### Train Test Split

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=50)
```

```python
X_train
```

```python
y_train
```

### Scaling

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

scaled_X_train = scaler.fit_transform(X_train)

scaled_X_test = scaler.transform(X_test)
```

## Modeling

### Logistic Regression Model

```python
from sklearn.linear_model import LogisticRegression

lr_model = LogisticRegression()

lr_model.fit(scaled_X_train , y_train)
```

```python
y_pred = lr_model.predict(scaled_X_test)
```

```python
from sklearn.metrics import accuracy_score

accuracy_score(y_test, y_pred)
```

### KNN Model

```python
from sklearn.neighbors import KNeighborsClassifier

knn_model = KNeighborsClassifier()

knn_model.fit(scaled_X_train, y_train)

```

```python
y_pred = knn_model.predict(scaled_X_test)
```

```python
accuracy_score(y_test, y_pred)
```

### SVM Model

```python
from sklearn.svm import SVC

svc_model = SVC(kernel = 'linear')

svc_model.fit(scaled_X_train, y_train)
```

```python
y_pred = svc_model.predict(scaled_X_test)
```

```python
accuracy_score(y_test, y_pred)
```

### Decision Tree Model

```python
from sklearn.tree import DecisionTreeClassifier

dt_model = DecisionTreeClassifier()

dt_model.fit(scaled_X_train, y_train)
```

```python
y_pred = dt_model.predict(scaled_X_test)
```

```python
accuracy_score(y_test, y_pred)
```

### Random Forest Model

```python
from sklearn.ensemble import RandomForestClassifier

rf_model = RandomForestClassifier()

rf_model.fit(scaled_X_train, y_train)
```

```python
y_pred = rf_model.predict(scaled_X_test)
```

```python
accuracy_score(y_test, y_pred)
```

### AdaBoost Model

```python
from sklearn.ensemble import AdaBoostClassifier

ada_model = AdaBoostClassifier()

ada_model.fit(scaled_X_train, y_train)
```

```python
y_pred = ada_model.predict(scaled_X_test)
```

```python
accuracy_score(y_test, y_pred)
```

### GradientBoost Model

```python
from sklearn.ensemble import GradientBoostingClassifier

gb_model = GradientBoostingClassifier()

gb_model.fit(scaled_X_train, y_train)
```

```python
y_pred = gb_model.predict(scaled_X_test)
```

```python
accuracy_score(y_test, y_pred)
```

### XGBoost Model

```python
from xgboost import XGBClassifier

xgb_model = XGBClassifier()
```

```python
xgb_model.fit(scaled_X_train, y_train)
```

```python
y_pred = xgb_model.predict(scaled_X_test)
```

```python
accuracy_score(y_test, y_pred)
```

### Hyperparameter Tuning

```python
from sklearn.model_selection import GridSearchCV
```

```python
estimator = LogisticRegression()

param_grid = {'C':[100,10,1.0,0.1,0.01],'penalty':['l2'], 'solver':['newton-cg','lbfgs','liblinear']}
```

```python
grid = GridSearchCV(estimator, param_grid, scoring='accuracy', cv=5, n_jobs=-1, error_score=0)
```

```python
grid.fit(scaled_X_train, y_train)
```

```python
grid.best_params_
```

### Final Model

```python
final_model = LogisticRegression(C=100, penalty='l2',solver='newton-cg')

final_model.fit(scaled_X_train, y_train)
```

```python
y_pred_test = final_model.predict(scaled_X_test)

y_pred_train = final_model.predict(scaled_X_train)
```

```python
print(accuracy_score(y_test, y_pred_test))

print(accuracy_score(y_train, y_pred_train))
```

### Saving a Model ( Serialization)

```python
from pickle import dump

dump(final_model, open('models/Diabetes_final_lr.pkl', 'wb'))
dump(xgb_model, open('models/xgboost.pkl', 'wb'))
dump(gb_model, open('models/gradientboost.pkl', 'wb'))
dump(ada_model, open('models/Adaboost.pkl', 'wb'))
dump(rf_model, open('models/Random_forest.pkl', 'wb'))
dump(dt_model, open('models/Decision_Tree.pkl', 'wb'))
dump(svc_model, open( 'models/SVM_model.pkl', 'wb'))
dump(knn_model, open('models/KNN_model.pkl','wb'))
dump(lr_model, open('models/Logistic_Regression.pkl', 'wb'))
dump(scaler, open('models/StandardScaler.pkl','wb'))
```
